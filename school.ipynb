```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5, 6, 7]
y = [10, 20, 15, 25, 30, 28, 35]

plt.plot(x, y, label="Class score")
plt.xlabel("Range")
plt.ylabel("Scores")
plt.title("Line Graph of Class Scores")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
from collections import Counter

marks = [12,11,19,12,19,10,11,13,14,10,15,10,11,13,12]
freq = Counter(marks)

plt.bar(freq.keys(), freq.values())
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.title("Frequency of Marks")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

names = ["Amit", "Renu", "Cheena"]
english = [20, 50, 30]
hindi = [25, 45, 20]

x = np.arange(len(names))
width = 0.35

plt.bar(x - width/2, english, width, label="English", color="red")
plt.bar(x + width/2, hindi, width, label="Hindi", color="blue")

plt.xlabel("Class")
plt.ylabel("Average")
plt.title("Comparative Chart")
plt.xticks(x, names)
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy import stats

data = [12,11,19,12,19,10,11,13,14,10,15,10,11,13,12]

mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data)

print("Mean:", mean)
print("Median:", median)
# Check if mode.mode and mode.count are arrays or scalars
if isinstance(mode.mode, np.ndarray):
    print("Mode:", mode.mode[0], "Frequency:", mode.count[0])
else:
    print("Mode:", mode.mode, "Frequency:", mode.count)
```
